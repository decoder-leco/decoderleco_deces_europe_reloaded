# Recuperation des données pour https://opendata.ecdc.europa.eu/covid19/vaccine_tracker/csv/data.csv

```typescript
const response: Promise = await fetch('https://opendata.ecdc.europa.eu/covid19/vaccine_tracker/csv/data.csv')
```

```typescript
const data: String = await await response.text();
```

## Nous utiliserons polars pour valider & visualiser nos donnéées

```typescript
import pl from "npm:nodejs-polars";
```

## extraction des types de chaques colonnes

```typescript
 pl.readCSV(data, { sep: "," }).dtypes
```

## Affichage des 100 1ers rangs

```typescript
 pl.readCSV(data, { sep: "," }).head(100)
```
