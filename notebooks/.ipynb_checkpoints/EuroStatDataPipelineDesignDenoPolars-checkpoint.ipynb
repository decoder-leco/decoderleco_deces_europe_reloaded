## Download Data

We download various datasets coming from https://ec.europa.eu/eurostat/web/main/data/database

For all those datasets,onlyone format is available, it seems: the `TSV` format

```typescript
const eurostat_baseUrl: String = "https://ec.europa.eu/eurostat/api/dissemination/sdmx/2.1/data/"
```

### Example 1: Animal populations by NUTS 2 regions (`tgs00045`)

* The data itself is at https://ec.europa.eu/eurostat/api/dissemination/sdmx/2.1/data/tgs00045?format=TSV&compressed=true
* The (very) detailed metadata, including things like measure units are available as an HTML page at https://ec.europa.eu/eurostat/cache/metadata/en/apro_anip_esms.htm

#### From TSV format

```typescript
// let desiredFormat: String = "/?format=SDMX-CSV"
// let desiredFormat: String = "/?format=CSV" // <?xml version="1.0" encoding="UTF-8"?><S:Fault xmlns:S="http://schemas.xmlsoap.org/soap/envelope/"><faultcode>140</faultcode><faultstring>UNSUPPORTED_FORMAT: Format is not supported</faultstring></S:Fault>
let desiredFormat: String = "/?format=TSV"
let response: Promise = await fetch( eurostat_baseUrl +"tgs00045"+ desiredFormat)
let animal_pop_by_nuts_2_regions_tgs00045: String = await response.text();
```

```typescript
console.log(animal_pop_by_nuts_2_regions_tgs00045)
```

```typescript
// pl.readCSV(animal_pop_by_nuts_2_regions_tgs00045.substring(0,demo_r_mwk_05.indexOf('\r\n')), { sep: ","});
pl.readCSV(animal_pop_by_nuts_2_regions_tgs00045, { sep: ","});
```

```typescript
import * as tsvJson from 'npm:tsv-json'
```

```typescript
import { tsv2json, json2tsv } from 'npm:tsv-json';

console.log(json2tsv([
    ['foo', 'bar'],
    ['baz', 'qux']
]));
//=> 'foo\tbar\nbaz\tqux'
 
console.log(tsv2json('foo\tbar\nbaz\tqux'));
//=> [['foo', 'bar'], ['baz', 'qux']]
```

```typescript


 
console.log(tsv2json(animal_pop_by_nuts_2_regions_tgs00045));
```

```typescript
import { json2csv } from 'npm:json-2-csv';

 
// console.log(tsv2json(animal_pop_by_nuts_2_regions_tgs00045));
const animal_pop_by_nuts_2_regions_tgs00045_as_csv = await json2csv(tsv2json(animal_pop_by_nuts_2_regions_tgs00045), {});
//console.log(animal_pop_by_nuts_2_regions_tgs00045_as_csv)


const df_animal_pop_by_nuts_2_regions_tgs00045 = pl.readCSV(animal_pop_by_nuts_2_regions_tgs00045_as_csv, { sep: "," });
df_animal_pop_by_nuts_2_regions_tgs00045
```

##### Cleaning the quotes in data

_the data headers are quoted in the CSV Files: we remove any " character from all frames data_

```typescript
const animal_pop_by_nuts_2_regions_tgs00045_as_csv_cleaned = animal_pop_by_nuts_2_regions_tgs00045_as_csv.replace(/"/g, "");

const df_animal_pop_by_nuts_2_regions_tgs00045_cleaned = pl.readCSV(animal_pop_by_nuts_2_regions_tgs00045_as_csv_cleaned, { sep: "," });
df_animal_pop_by_nuts_2_regions_tgs00045_cleaned
```

### Example 2: Population on 1 January by age and sex (`demo_pjan`)

* The data itself is at https://ec.europa.eu/eurostat/api/dissemination/sdmx/2.1/data/demo_pjan?format=TSV&compressed=true
* The (very) detailed metadata, including things like measure units are available as an HTML page at https://ec.europa.eu/eurostat/cache/metadata/en/demo_pop_esms.htm



```typescript
let desiredFormat: String = "/?format=TSV"
const response_Population_on_1_January_by_age_and_sex_demo_pjan: Promise = await fetch( eurostat_baseUrl +"demo_pjan"+ desiredFormat)
const Population_on_1_January_by_age_and_sex_demo_pjan: String = await response_Population_on_1_January_by_age_and_sex_demo_pjan.text();
// console.log(Population_on_1_January_by_age_and_sex_demo_pjan);
```

```typescript
/**
 * we thenwrite the string data to a file, because its much too big to be plaayed with inside a single variable in RAM
 * After that, we willlook for ways to split the TSV data into multiple files, that can be
 * processed independently, to be turned into JSON for example ?
 **/
// await Deno.writeTextFile("Population_on_1_January_by_age_and_sex_demo_pjan.tsv", Population_on_1_January_by_age_and_sex_demo_pjan);
```

```typescript
const tsv2arr = (tsv) => {
  const [headers, ...rows] = tsv .trim () .split ('\n') .map (r => r .split ('\t'))
  return rows .reduce ((a, r) => [
    ... a, 
    Object .assign (... (r .map (
      (x, i, _, c = x.trim()) => ({[headers [i].trim()]: isNaN(c) ? c : Number(c)})
    )))
  ], [])
}

const MyResultingJSON = { array: tsv2arr(Population_on_1_January_by_age_and_sex_demo_pjan) }
console.log (`MyResultingJSON`,
  MyResultingJSON
)
```

```typescript
// const Population_on_1_January_by_age_and_sex_demo_pjan_as_json = await tsv2json(Population_on_1_January_by_age_and_sex_demo_pjan);
const Population_on_1_January_by_age_and_sex_demo_pjan_as_json = MyResultingJSON.array;
// console.log(Population_on_1_January_by_age_and_sex_demo_pjan_as_json)
```

```typescript
//const Population_on_1_January_by_age_and_sex_demo_pjan_as_csv = await json2csv(tsv2json(Population_on_1_January_by_age_and_sex_demo_pjan), {});
const Population_on_1_January_by_age_and_sex_demo_pjan_as_csv = await json2csv(Population_on_1_January_by_age_and_sex_demo_pjan_as_json, {});
// console.log(Population_on_1_January_by_age_and_sex_demo_pjan_as_csv)
```

```typescript

const Population_on_1_January_by_age_and_sex_demo_pjan_as_csv_cleaned = Population_on_1_January_by_age_and_sex_demo_pjan_as_csv.replace(/"/g, "");
```

```typescript

const df_Population_on_1_January_by_age_and_sex_cleaned = pl.readCSV(Population_on_1_January_by_age_and_sex_demo_pjan_as_csv_cleaned, { sep: "," });
df_Population_on_1_January_by_age_and_sex_cleaned
```
